```
from mimesis import Person
from mimesis import Address
from mimesis.enums import Gender
from mimesis import Datetime
import pandas as pd
import random
from mimesis import locales

person = Person('en')
addess = Address(locales.EN_AU)
datetime = Datetime()
def create_rows_mimesis(num=1):
    output = [{"name":person.full_name(gender=Gender.FEMALE),
                   "address":addess.address(),
                   "name":person.name(),
                   "email":person.email(),
                   "blood_type":person.blood_type(),
                   "city":addess.city(),
                   "state":addess.state(),
                   "last_login":datetime.datetime(),
                   "height":person.height(),
                   "weight":person.weight(), 
                   "random_data":random.randint(1000,100000+1000)} for x in range(num)]
    return output
```

```
%%time
df_mimesis = pd.DataFrame(create_rows_mimesis(15000))
```

```
df_mimesis.state = df_mimesis.state.astype('category')
df_mimesis.city = df_mimesis.city.astype('category')
df_mimesis.blood_type = df_mimesis.blood_type.astype('category')
df_mimesis.height = pd.to_numeric(df_mimesis.height, errors='coerce')
```

```
df_mimesis.index.set_names('unique_code',inplace=True)
```

```
df_mimesis = df_mimesis.reset_index()
```

```
df_citizens = df_mimesis[['unique_code','name','email','last_login']]
df_health_care = df_mimesis[['unique_code','blood_type','height','weight']]
df_address = df_mimesis[['unique_code','city','address','state','random_data']]
```

```
from snowflake.sqlalchemy import URL
from sqlalchemy import create_engine
engine = create_engine(URL(
    user='romellfudi',
    password='freddyFUDI9568',
    account='gr49894.west-us-2.azure',
    warehouse = 'boosttag_dw',
    database = 'boosttag_db', 
    schema = 'australian_sc',
    role='SYSADMIN',
))
 
connection = engine.connect()
 
df_citizens.to_sql('df_citizens', con=engine, index=False, if_exists='replace') #make sure index is False, Snowflake doesnt accept indexes
df_health_care.to_sql('df_health_care', con=engine, index=False, if_exists='replace')
df_address.to_sql('df_address', con=engine, index=False, if_exists='replace')
 
connection.close()
engine.dispose()
```

```
import snowflake.connector
import pandas as pd

# Gets the version
ctx = snowflake.connector.connect(
    user='romellfudi',
    password='freddyFUDI9568',
    account='gr49894.west-us-2.azure',
    role='SYSADMIN',
    )
cs = ctx.cursor()
try:
    cs.execute("SELECT * FROM boosttag_db.australian_sc.DF_HEALTH_CARE;")
    health_care = pd.DataFrame(cs.fetchall(),columns=list(zip(*cs.description))[0])
    cs.execute("SELECT * FROM boosttag_db.australian_sc.DF_ADDRESS;")
    address = pd.DataFrame(cs.fetchall(),columns=list(zip(*cs.description))[0])
    cs.execute("SELECT * FROM boosttag_db.australian_sc.DF_CITIZENS;")
    citizens = pd.DataFrame(cs.fetchall(),columns=list(zip(*cs.description))[0])
finally:
    cs.close()
ctx.close()
```
